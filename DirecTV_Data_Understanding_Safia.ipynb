# DirecTV Data Understanding

```python
import pandas as pd
import os
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme() # this line activates a signature aesthetic that makes seaborn plots look better
```

#### Load a Data Set and Save it as a Pandas DataFrame

```python
filename = os.path.join(os.getcwd(), "original_tv_show_dataset_2024.csv")
df = pd.read_csv(filename)
```

#### 1. Basic Exploration of IMDB Data

* Basic Information

```python
# display the first few rows of the DataFrame
display(df.head())
```

```python
# display the last few rows of the DataFrame
display(df.tail())
```

```python
# display basic statistics of numeric columns
display(df.describe())
```

```python
# display information about DataFrame
display(df.info())
```

```python
# display the most occuring original languages
original_language_counts = df["original_language"].value_counts()

with pd.option_context("display.max_rows", None):
  display(original_language_counts)
```

```python
# display the shows with the highest vote average
sorted_shows_by_vote_average = df[["name", "vote_count", "vote_average"]].sort_values(by="vote_average", ascending=False)

display(sorted_shows_by_vote_average)
```

```python
# display shows that have both a high vote count and a high vote average

high_vote_average_threshold = 7
high_vote_count_threshold = 100

filtered_shows = df[(df["vote_average"] >= high_vote_average_threshold) &
                    (df["vote_count"] >= high_vote_count_threshold)]

sorted_filtered_shows = filtered_shows[["name", "vote_average", "vote_count"]].sort_values(
    by=["vote_average", "vote_count"], ascending=False)

display(sorted_filtered_shows)

total_shows = len(df)
filtered_shows_count = len(filtered_shows)

# percentage of shows with over 100 votes
percentage = (filtered_shows_count / total_shows) * 100
print(f"{percentage:.2f}% of shows have over {high_vote_count_threshold} votes")

# percentage of shows with only 1 vote or 0 votes
shows_with_zero_or_one_vote = df[df["vote_count"].isin([0, 1])]
number_of_shows_with_zero_or_one_vote = len(shows_with_zero_or_one_vote)
percentage_of_shows_with_zero_or_one_vote = (number_of_shows_with_zero_or_one_vote / total_shows) * 100

print(f"{percentage_of_shows_with_zero_or_one_vote:.2f}% of shows have either 0 or 1 votes")
```

```python
# display the most occuring genres
genres_counts = df["genres"].value_counts()

with pd.option_context("display.max_rows", None):
  display(genres_counts)
```

#### 2. Data Cleaning

* Handle Missing Values (Will decide as a group how to handle the missing data)

```python
# store the names of the columns with missing values
condition = nan_count != 0 # look for all columns with missing values

col_names = nan_count[condition].index # get the column names

nan_cols = list(col_names) # convert column names to list
print(f"Column Names:\n {nan_cols}")
```

```python
# data types of columns with missing values
nan_col_types = df[nan_cols].dtypes
nan_col_types
```

```python
# print the percentage of missing data within each column
missing_data = df.isnull().mean().sort_values(ascending=False)
print(missing_data)
```

#### 4. Bivariate Analysis

```python
# define a threshold to categorize vote_average
threshold = 7.0
df['vote_category'] = df['vote_average'].apply(lambda x: 'Above 7.0' if x >= threshold else 'Below 7.0')

# plotting original_language and vote_average
fig = plt.figure(figsize=(40,14))
sns.histplot(data=df, x='original_language', hue='vote_category', multiple='stack')
plt.xticks(rotation=45)
plt.xlabel('Original Language')
plt.ylabel('Count')
plt.title('Distribution of Shows by Original Language and Vote Average')
plt.show()
```
